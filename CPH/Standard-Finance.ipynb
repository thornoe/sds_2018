```python
#Kontant udbetaling på 5% af den kontante købesum oprundet til nærmeste kr. 5.000 dog minimum kr. 25.000
Houseprice = 30012000

rem = Houseprice % 5000
if rem < 2500:
    Houseprice = int(Houseprice / 5000) * 5000
else:
    Houseprice = int((Houseprice + 5000) / 5000) * 5000

Cashticket = max(Houseprice*0.05,25000)
print(Cashticket)


# Vi antager at køberne selvfinansiere de 20 % af købssummen.
# Mens de låner 80 % i et realkreditinstituet
# FIRST: vi antager der er en ÅOP på 3% efter skat på realkreditlånet.
# SECOND: Vi antager de låner de resterende 15 porocent i banken til en ÅOP på 6.6% efter skat banklånet


```

```python
#Selvfinansiering:
'''U = udbetaling'''
Owner_expense = 12000
Houseprice = 3000000
Bankloan = ((Houseprice - Cashticket)*0.15)*0.066
Mortgage = (Houseprice - Bankloan)*0.03
Yearly_expense = Owner_expense + Bankloan + Mortgage
first_year_expense = Owner_expense + Bankloan + Mortgage + Cashticket
print(str(Yearly_expense) + ' DKK', str(first_year_expense) + ' DKK')
```
